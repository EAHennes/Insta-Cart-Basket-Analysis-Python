```python
#creating flags for price ranges
df=df_merged[:1000000]
```

```python
#creating a new row by defining it 
def price_label(row):
    
    if row ['prices']<= 5 :
        return'Low-range product'
    elif(row['prices']>5) and (row['prices']<= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
#now that row is created I need to apply it to the DF like a merege 
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts (dropna = False)
```

```python
#using loc function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices']<= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid ranged product'
```

```python
df.loc[df['prices'] <=5, 'price_range_loc'] = 'Low range product'
```

```python
result = []

for value in df_merged["orders_day_of_week"]:
    if value == 0:
        result.append("Most busy day")
    elif value == 1:
        result.append("Second most busy day")
    elif value == 4:
        result.append("Least busy day")
    elif value == 3:
        result.append("Second least busy day")
    else:
        result.append("Regularly busy")
```
