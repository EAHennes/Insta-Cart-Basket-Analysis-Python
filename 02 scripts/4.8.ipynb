```python
#6) To create a new flag and aggregation to lable users as "low spending" and high "spending" my steps are as follows
#a) I will create a new group/column for the mean amount spent per customer
#b) Then I will use this new column to create a for loop
```

```python
df_ords_prods_merge['average_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge[['user_id','prices','average_price']].head(100)
```

```python
#creating a for loop for the new column
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] < 10,'price_flag'] = "Low spender"
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] >10,'price_flag']= "High Spender"
```

```python
#running some checks
df_ords_prods_merge['price_flag'].value_counts(dropna=False)
```

```python
df_ords_prods_merge[['user_id','price_flag','average_price']].head(60)
```

```python
#7) Following the same steps as above to aggreagte and create a flag for frequency of use
df_ords_prods_merge['average_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_from_last_order'].transform(np.mean)
```

```python
#Checking head
df_ords_prods_merge.head()
```

```python
#creating for new criteria for aggregation
df_ords_prods_merge.loc[df_ords_prods_merge['average_frequency']>20,'frequency_flag'] = "Non-frequent Customer"
df_ords_prods_merge.loc[(df_ords_prods_merge['average_frequency']>10) & (df_ords_prods_merge['average_frequency']<= 20), 'frequency_flag'] = 'regular_customer'
df_ords_prods_merge.loc[df_ords_prods_merge['average_frequency']<=10,'frequency_flag'] = 'frequent Customer'                                             

```

```python
#running a check on new aggregation
df_ords_prods_merge[['user_id','frequency_flag','average_frequency']].head(60)
```
